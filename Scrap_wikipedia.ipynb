```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
page = requests.get(url)
```

```python
page dvdvdfvdvdf
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
soup.find_all('table')[1]
```

```python
result = soup.find("table", class_ = "wikitable sortable")
```

```python
print(result.prettify())
```

```python
world_titles = result.find_all('th')[1:]
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles[1:]]
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = result.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')[1:]
    invidual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = invidual_row_data

df
```
